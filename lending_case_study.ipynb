## Defaulters dataframe

```python
default_loan_df=loan_df.loc[loan_df['loan_status']=='Charged Off']
default_loan_df.head(10)
```

## calculate the percentage of loan defaulters according to term

```python
term_df=pd.DataFrame(data=round((100*default_loan_df['term'].value_counts())/loan_df['term'].value_counts(),2))
term_df.rename(columns={'term':'defaulter_percentage'},inplace=True)
term_df.sort_values(by='defaulter_percentage',ascending=False,inplace=True)
term_df
```

## 60 months term defaulters more than 36 months

```python
sns.barplot(x=term_df.index, y='defaulter_percentage', hue=term_df.index, data=term_df, palette='Oranges', legend=True)
plt.xlabel('Loan Term in Months')
plt.ylabel('Percentage')
plt.title('Percentage by Loan Term')
plt.legend(title='Loan Term in Months')
plt.show()
```

## Grade wise ChargedOff loans

```python
grade_order = default_loan_df['grade'].value_counts().sort_index()
plt.figure(figsize=(16, 6))
sns.countplot(x='grade', data=default_loan_df, legend=False, order=grade_order.index)
plt.xlabel('Grades')
plt.ylabel('Charged_Off loans')
plt.title('Percentage of Sub Grades for Charged_Off status')
plt.show()
```

## Sub grade wise charged off loans

```python
grouped_df = loan_df.groupby(['sub_grade', 'loan_status']).size().reset_index(name='count')

# Calculating percentage based on sub_grade counts
sub_grade_counts = loan_df['sub_grade'].value_counts().reset_index()
sub_grade_counts.columns = ['sub_grade', 'sub_grade_count']

grouped_df = pd.merge(grouped_df, sub_grade_counts, on='sub_grade')
grouped_df['perc'] = 100 * grouped_df['count'] / grouped_df['sub_grade_count']
grouped_df=grouped_df[grouped_df['loan_status'] == 'Charged Off']
grouped_df = grouped_df.sort_values(by='perc', ascending=True)
grouped_df
```

## Sub Grade A1 is better than A2 and A3

```python
plt.figure(figsize=(16, 6))
sns.barplot(x='sub_grade', y='perc', hue='sub_grade', data=grouped_df, legend=False)
plt.xlabel('Sub Grades')
plt.ylabel('Sub Grades Percentage for Charged_Off status')
plt.title('Percentage of Sub Grades for Charged_Off status')
plt.show()
```

```python
# chaning emp_length column with appropriate numbers
loan_df['emp_length'] = loan_df['emp_length'].astype("str")
loan_df['emp_length'] = loan_df['emp_length'].apply(lambda x: x.replace("years",""))
loan_df['emp_length'] = loan_df['emp_length'].apply(lambda x: x.replace("year",""))
loan_df['emp_length'] = loan_df['emp_length'].apply(lambda x: x.replace("10+","11"))
loan_df['emp_length'] = loan_df['emp_length'].apply(lambda x: x.replace("< 1","0"))
loan_df.head(10)
```

## Employee length dependency on loan status

```python
loan_df['emp_length'].value_counts()
```

```python
#removing nan/ null values from emp_length
loan_df = loan_df[~(loan_df['emp_length']=="nan")]
loan_df['emp_length'].value_counts()
loan_df['emp_length'].value_counts()
```

```python
grouped_emp_length_df = loan_df.groupby(['emp_length', 'loan_status']).size().reset_index(name='count')

# Calculating percentage based on sub_grade counts
emp_length_counts = loan_df['emp_length'].value_counts().reset_index()
emp_length_counts.columns = ['emp_length', 'emp_length_counts']

grouped_emp_length_df = pd.merge(grouped_emp_length_df, emp_length_counts, on='emp_length')
grouped_emp_length_df['perc'] = 100 * grouped_emp_length_df['count'] / grouped_emp_length_df['emp_length_counts']
grouped_emp_length_df=grouped_emp_length_df[grouped_emp_length_df['loan_status'] == 'Charged Off']
grouped_emp_length_df = grouped_emp_length_df.sort_values(by='perc', ascending=True)
grouped_emp_length_df
```

```python
#Employment length: not much difference on the defaulter
plt.figure(figsize=(10, 6))
sns.barplot(x='emp_length', y='perc', hue='emp_length', data=grouped_emp_length_df, legend=False)
plt.xlabel('Employee length in Years')
plt.ylabel('Percentage')
plt.title('Percentage by Loan Status Charged off')
plt.show()
```
